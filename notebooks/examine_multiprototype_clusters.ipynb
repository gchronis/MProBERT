```python
word = 'lion'


def get_sentences_in_cluster(sentence_uids, word):
    tokens = []

    with open('../data/word_data/'+word+'/BNC_tokens.csv', mode="r") as infile:
        fieldnames = ['word', 'sentence', 'tag', 'bnc_index']
        reader = csv.DictReader(infile, delimiter="\t", fieldnames=fieldnames)
        
        for row in reader:
            if row['bnc_index'] in sentence_uids:
                tokens.append(row['sentence'])
    return tokens
        
#pickle_file = '../data/word_data/'+word+'/analysis_results/clusters.p'
#data = pickle.load(pickle_file)
 
clusters = helpers.read_clusters(word)
clusters = clusters[clusters['k_clusters'] == 7]
clusters = clusters[clusters['layer'] == 8]

print(clusters)

# for each of the clusters, go through the tokens we hae for that word and select the tokens that match the sentence uid
for index, cluster in clusters.iterrows():
    sentence_uids = cluster['sentence_uids']
    tokens = get_sentences_in_cluster(sentence_uids, word)
    print(cluster['cluster_id'])
    for token in tokens:
        print(token)
```

```python
#compare the clusters of bishop and rook and rabbi

k = 7
l = 8

def cluster_centroids_at_layer_and_k(word, l, k):
    clusters = helpers.read_clusters(word)
    clusters = clusters[clusters['k_clusters'] == k]
    clusters = clusters[clusters['layer'] == l]
    return clusters
    

# bishop vectors
bishop = cluster_centroids_at_layer_and_k('cat', l, k)
print(bishop)

# rook vectors
rook = cluster_centroids_at_layer_and_k('lion', l, k)

# rabbi vectors
#rabbi = cluster_centroids_at_layer_and_k('rabbi', l, k)

print("bishop-rook")
for i in bishop.index:
    for j in rook.index:
        print("word1 cluster = %s, word2 cluster = %s" % (bishop['cluster_id'][i],rook['cluster_id'][j]))
        sim = 1- cosine(bishop['centroid'][i],rook['centroid'][j])
        print(sim)
```
