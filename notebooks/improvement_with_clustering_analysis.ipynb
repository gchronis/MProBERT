So you have seen that clustering leads to a marked improvement in perfoemance on word similarity tasks like SimLex999.

But I wonder. Is the improvement wholesale? Are there some particular word-pairs/relations or types of relations for which it is improving drastically? Or is there a general improvement across the board?

Here is where we can dig down a little deeper to determine where the improvements are being made.


First we will look at a particular layer. Layer 5. Where we get the best overall performance with 3 clusters.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import spearmanr
from scipy.spatial.distance import cosine
import os
import csv
import pickle

from helpers import datasets


DATA_DIR = '../data/word_data'

def read_centroids_for_word_at_layer_and_cluster(word, layer_number, k):
    try:
        cluster_path = os.path.join(DATA_DIR, word, 'analysis_results', 'clusters.p')
        """
         this is a list of dicts with the structure:
                        {'word': tokens[0]['word'],
                        'layer': layer,
                        'k_clusters': k,
                        'cluster_id': cluster_index,
                        'centroid': cluster_centroids[cluster_index],
                        'sentence_uids': sentence_uids,
                        'within_cluster_variance': cluster_variance
                        }
        """
        data = pickle.load(open(cluster_path, 'rb'))
        data = [item for sublist in data for item in sublist]

    except:
        return None

    columns = ['word', 'layer', 'k_clusters', 'cluster_id', 'centroid', 'sentence_uids', 'within_cluster_variance']
    df = pd.DataFrame.from_records(data, columns=columns)
    #print(df)

    df = df[df['layer'] == layer_number]
    df = df[df['k_clusters'] == k]
    word_centroids = df['centroid']
    if len(word_centroids) > 0:
        return word_centroids
    else:
        return None


def max_sim(w1_centroids, w2_centroids):
    predicted_similarities = []
    for centroid1 in w1_centroids:
        for centroid2 in w2_centroids:
            predicted_similarity = 1 - cosine(centroid1, centroid2)
            # TODO am I nuts? Why do I get a positive correlation for cosine distance over cosine similarity???
            #predicted_similarity = cosine(centroid1, centroid2)
            predicted_similarities.append(predicted_similarity)
    # find the max of the pairwise similarities
    return max(predicted_similarities)

"""
1) the words we want to collect data for
"""
ws353 = datasets.get_ws353()
simlex999 = datasets.get_simlex999()

"""
2) parameters we care about
"""
layer_number = 8
cluster_sizes = [2,3,4,5,6,7,8,9,10,50]

for k in cluster_sizes:

    # calc sim for all the word pairs
    data = simlex999
    expected_similarities = []
    for row in data:
        
        word1 = row['word1']
        word2 = row['word2']
        observed_similarity = row['similarity']
        scaled_observed_similarity = observed_similarity / 10

        """
        get the difference of predicted - observed for both 0 clusters and current num clusters
        """
        
        # get centroid data for these words at this layer and no clusters
        w1_centroids = read_centroids_for_word_at_layer_and_cluster(word1, layer_number, 1)
        w2_centroids = read_centroids_for_word_at_layer_and_cluster(word2, layer_number, 1)
        # find the max of the pairwise similarities
        if w1_centroids is not None and w2_centroids is not None:
            predicted = max_sim(w1_centroids, w2_centroids)
            scaled_predicted = predicted / 2
            # calculate 'score' for this word pair
            no_cluster_difference = scaled_predicted - scaled_observed_similarity
            row['no_cluster_difference'] = no_cluster_difference
        else:
            row['no_cluster_difference'] = None
        
        # get centroid data for these words at this layer and this k size
        w1_centroids = read_centroids_for_word_at_layer_and_cluster(word1, layer_number, k)
        w2_centroids = read_centroids_for_word_at_layer_and_cluster(word2, layer_number, k)
        
        
        row['num_clusters'] = k
        #do you have both centroid lists? don't proceed if you dont havve enough data points
        if w1_centroids is not None and w2_centroids is not None:
            # find the max of the pairwise similarities
            predicted = max_sim(w1_centroids, w2_centroids)
            scaled_predicted = predicted / 2
            # calculate 'score' for this word pair
            cluster_difference = scaled_predicted - scaled_observed_similarity
            row['cluster_difference'] = cluster_difference
            
            """
            improvement_with_clusters = |no_cluster_difference| - |cluster_difference|
            this will give you a positive value if you have made an improvement 
            (if the new prediction is closer to the gold similarity label than the old prediction)
            and a negative value if you have declined.
            """
            improvement_with_clusters = abs(no_cluster_difference) - abs(cluster_difference)
            row['improvement'] = improvement_with_clusters
            row['scaled_predicted'] = scaled_predicted
            row['scaled_observed'] = scaled_observed_similarity
        else:
            row['cluster_difference'] = None
            row['improvement'] = None
            row['scaled_predicted'] = None
            row['scaled_observed'] = scaled_observed_similarity
 
        
    data = list(filter(lambda row: row['cluster_difference'] != None, data))

    columns = ['num_clusters', 'word1', 'word2', 'improvement', 'POS', 'SimLex999', 'assoc_USF', 'scaled_predicted', 'scaled_observed']
    df = pd.DataFrame.from_records(data, columns=columns)

    print("num clusters %s" % k)
    #print(df)
    df['improvement'].hist(bins=40)
    plt.show()
    
    print("word pairs which improved")
    count_by_pos = {'N': 0, 'A': 0, 'V': 0}
    percent_by_pos = {'N': 666, 'A': 111, 'V': 222}
    sub_df = df[df['improvement'] > 0.0]
    for pos in ('N', 'A', 'V'):
        count_by_pos[pos] = len(sub_df[sub_df['POS'] == pos])
        percent_by_pos[pos] = round ( len(sub_df[sub_df['POS'] == pos]) / percent_by_pos[pos] , 3)
    print(count_by_pos)
    print(percent_by_pos)
    
    print("word pairs which degraded")
    count_by_pos = {'N': 0, 'A': 0, 'V': 0}
    percent_by_pos = {'N': 666, 'A': 111, 'V': 222}
    sub_df = df[df['improvement'] < 0.0]
    for pos in ('N', 'A', 'V'):
        count_by_pos[pos] = len(sub_df[sub_df['POS'] == pos])
        percent_by_pos[pos] = round ( len(sub_df[sub_df['POS'] == pos]) / percent_by_pos[pos] , 3)
    print(count_by_pos)
    print(percent_by_pos)
    
    print("word pairs which improved < 0.05:")
    count_by_pos = {'N': 0, 'A': 0, 'V': 0}
    percent_by_pos = {'N': 666, 'A': 111, 'V': 222}
    sub_df = df[df['improvement'] < 0.05][df['improvement'] > 0.0]
    #print(sub_df)
    for pos in ('N', 'A', 'V'):
        count_by_pos[pos] = len(sub_df[sub_df['POS'] == pos])
        percent_by_pos[pos] = round ( len(sub_df[sub_df['POS'] == pos]) / percent_by_pos[pos] , 3)
    print(count_by_pos)
    print(percent_by_pos)
    
    print("word pairs which improved > 0.05:")
    count_by_pos = {'N': 0, 'A': 0, 'V': 0}
    percent_by_pos = {'N': 666, 'A': 111, 'V': 222}
    sub_df = df[df['improvement'] > 0.05]
    #print(sub_df)
    for pos in ('N', 'A', 'V'):
        count_by_pos[pos] = len(sub_df[sub_df['POS'] == pos])
        percent_by_pos[pos] = round ( len(sub_df[sub_df['POS'] == pos]) / percent_by_pos[pos] , 3)
    print(count_by_pos)
    print(percent_by_pos)
    
    print("word pairs which improved > 0.10:")
    count_by_pos = {'N': 0, 'A': 0, 'V': 0}
    percent_by_pos = {'N': 666, 'A': 111, 'V': 222}
    sub_df = df[df['improvement'] > 0.1]
    print(sub_df)
    for pos in ('N', 'A', 'V'):
        count_by_pos[pos] = len(sub_df[sub_df['POS'] == pos])
        percent_by_pos[pos] = round ( len(sub_df[sub_df['POS'] == pos]) / percent_by_pos[pos] , 3)
    print(count_by_pos)
    print(percent_by_pos)
    
    print("word pairs which degraded < 0.05:")
    count_by_pos = {'N': 0, 'A': 0, 'V': 0}
    percent_by_pos = {'N': 666, 'A': 111, 'V': 222}
    sub_df = df[df['improvement'] > -0.05][df['improvement'] < 0.0]
    #print(sub_df)
    for pos in ('N', 'A', 'V'):
        count_by_pos[pos] = len(sub_df[sub_df['POS'] == pos])
        percent_by_pos[pos] = round ( len(sub_df[sub_df['POS'] == pos]) / percent_by_pos[pos] , 3)
    print(count_by_pos)
    print(percent_by_pos)
    
    print("word pairs which degraded > 0.05:")
    count_by_pos = {'N': 0, 'A': 0, 'V': 0}
    percent_by_pos = {'N': 666, 'A': 111, 'V': 222}
    sub_df = df[df['improvement'] < -0.05]
    print(sub_df)
    for pos in ('N', 'A', 'V'):
        count_by_pos[pos] = len(sub_df[sub_df['POS'] == pos])
        percent_by_pos[pos] = round ( len(sub_df[sub_df['POS'] == pos]) / percent_by_pos[pos] , 3)
    print(count_by_pos)
    print(percent_by_pos)

```
